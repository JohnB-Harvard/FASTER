```python
# For Harris Corners
harris_imgs = []
for dir in dirs:
    print(f"Running {dir}")
    images = glob.glob(f"{dir}/events*.png")
    img = cv2.imread(images[0], cv2.IMREAD_GRAYSCALE)

    def compute_harris(img):
        harrisCorners = cv2.cornerHarris(img, blockSize=2, ksize=3, k=.04)
        harrisCorners = cv2.dilate(harrisCorners,None)
        ret, dst = cv2.threshold(harrisCorners,0.01*harrisCorners.max(),255,0)
        dst = np.uint8(dst)
        # find centroids
        ret, labels, stats, centroids = cv2.connectedComponentsWithStats(dst)
        return centroids
    %timeit compute_harris(img)
    centroids = compute_harris(img)
    imgH = cv2.imread(images[0], cv2.IMREAD_GRAYSCALE)
    imgH_draw = cv2.imread(images[0])
    for item in centroids:
        x, y = item 
        x = int(x) 
        y = int(y) 
        cv2.circle(imgH_draw, (x, y), 4, (0, 255, 0), 2) 
    harris_imgs.append(imgH_draw)
```

```python
for i, img in enumerate(harris_imgs):
    status = cv2.imwrite(f"{dirs[i]}\\Harris_2_results.png", img)
```

```python
# ORB
ORB_images = []
orb = cv2.ORB_create()

bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

def compute_orb(detector, img):
    kp = detector.detect(img,None)
    kp, des = detector.compute(img, kp)
    return kp, des

for dir in dirs:
    print(f"Running {dir}")
    images = glob.glob(f"{dir}/events*.png")
    img1 = cv2.imread(images[0], cv2.IMREAD_GRAYSCALE)
    img2 = cv2.imread(images[1], cv2.IMREAD_GRAYSCALE)
    # timing finding descriptors
    kp1, des1 = orb.detectAndCompute(img1, None)
    kp2, des2 = orb.detectAndCompute(img2, None)
    # Match descriptors.
    %timeit bf.match(des1,des2)
    matches = bf.match(des1,des2)
    # Sort them in the order of their distance.
    matches = sorted(matches, key = lambda x:x.distance)
    orb_img = cv2.drawMatches(img1,kp1,img2,kp2,matches[:10],None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

    ORB_images.append(orb_img)
```

```python
for i, img in enumerate(ORB_images):
    status = cv2.imwrite(f"{dirs[i]}\\ORB_Matching_Results.png", img)
```

```python
# FASTER (new solution)
def fast_corner_finder(image_a, px, py):
    circle = [(0,3), (1,3), (2,2), (3,1), (3,0), (3,-1), (2,-2), (1,-3), (0,-3), (-1,-3), (-2,-2), (-3,-1), (-3,0), (-3,1), (-2,2), (-1,3)]
    half_size = 3

    # This part of the method assumes a threshold of n=12 and we use n=9
    # count_pos = 0
    # count_neg = 0
    # for i in range(0,len(circle), int(len(circle)/4)):
    #     if image[py,px] > image[py+circle[i][0]][px+circle[i][1]]:
    #         count_neg += 1
    #     elif image[py,px] < image[py+circle[i][0]][px+circle[i][1]]:
    #         count_pos += 1
    # if(count_pos < 3 and count_neg < 3):
    #     return False 

    if py + half_size > len(image_a)-1 or px + half_size > len(image_a[0])-1 or py < half_size or px < half_size:
        return False
    if (image_a[py,px] == 0):
        return False
    if image_a[py-3,px] == image_a[py+3,px] and image_a[py,px-3] == image_a[py,px+3]:
        return False
    start_run = 0  
    start_run_ended = False
    max_run = 0
    cur_run = 0
    greater = image_a[py,px] < image_a[py+circle[0][0], px+circle[0][1]]
    greater_found = False
    less_found = False            
    for i in range(len(circle)):
        if image_a[py+circle[i][0], px+circle[i][1]] == -1:
            less_found = True
        if image_a[py+circle[i][0], px+circle[i][1]] == 1:
            greater_found = True
        if greater:
            if image_a[py,px] < image_a[py+circle[i][0], px+circle[i][1]]:
                cur_run += 1
                continue
            # end of current run
            if cur_run > max_run:
                max_run = cur_run
            if not start_run_ended:
                start_run = cur_run
                start_run_ended = True 
            if image_a[py,px] > image_a[py+circle[i][0], px+circle[i][1]]:
                cur_run = 1
                greater = False
            if image_a[py,px] == image_a[py+circle[i][0], px+circle[i][1]]:
                cur_run = 0

        else:
            if image_a[py,px] > image_a[py+circle[i][0], px+circle[i][1]]:
                cur_run += 1
                continue
            # end of current run
            if cur_run > max_run:
                max_run = cur_run
            if not start_run_ended:
                start_run = cur_run
                start_run_ended = True 
            if image_a[py,px] < image_a[py+circle[i][0], px+circle[i][1]]:
                cur_run = 1
                greater = True
            if image_a[py,px] == image_a[py+circle[i][0], px+circle[i][1]]:
                cur_run = 0
    if(cur_run + start_run > max_run):
        max_run = cur_run + start_run

    if (not (greater_found and less_found)):
        return False

    return max_run > int(len(circle)/2) and max_run != len(circle)

def compute_features(img, max_features, half_size, num_divs):
    feature_list = []
    all_failed = False
    while(len(feature_list) < max_features):
        all_failed = True
        for iy in range(num_divs):
            for ix in range(num_divs):
                region_found = False
                for py in range(iy*int(len(img_a)/num_divs), (iy+1)*int(len(img_a)/num_divs)):
                    for px in range(ix*int(len(img_a[0])/num_divs), (ix+1)*int(len(img_a[0])/num_divs)):
                        if(fast_corner_finder(img_a, px, py)):
                            found = False
                            for u in range(-half_size,half_size+1):
                                for v in range(-half_size,half_size+1):
                                    if (py+u,px+v) in feature_list:
                                        found = True
                                        break
                                if found:
                                    break
                            if not found:
                                region_found = True
                                all_failed = False
                                feature_list.append((py,px))
                        if region_found:
                            break
                    if region_found:
                        break
        if all_failed:
            break
    return feature_list

max_features = 36
half_size = 25
num_divs = 6

annotated_imgs = []
for dir in dirs:
    print(f"Running {dir}")
    images = glob.glob(f"{dir}/events*.png")
    img = cv2.imread(images[0], cv2.IMREAD_GRAYSCALE)
    img_draw = cv2.imread(images[0])
    img_a = np.array(img, dtype=np.float32)
    img_a = (img_a-128)/127
    print(np.amax(img_a))
    print(np.amin(img_a))
    %timeit compute_features(img_a, max_features, half_size, num_divs)
    feature_list = compute_features(img_a, max_features, half_size, num_divs)
    for feature in feature_list:
        cv2.circle(img_draw, (feature[1], feature[0]), 4, (0, 255, 0), 2)
    annotated_imgs.append(img_draw)

for i, img in enumerate(annotated_imgs):
    status = cv2.imwrite(f"{dirs[i]}\\FASTER_results.png", img)
```
